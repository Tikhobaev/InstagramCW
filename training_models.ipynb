## Oversampling

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.25)
X = pd.concat([X_train, y_train], axis=1)
no_fake = X[X.label == 0]
fake = X[X.label == 1]

fake_upsampled = resample(fake,
                          replace=True,
                          n_samples=len(no_fake))
X = pd.concat([fake_upsampled, no_fake])
X = X.sample(frac=1).reset_index(drop=True)
X_train = X.drop(['label'], axis=1)
y_train = X.label
X_train
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import *


def evalulate_classifier(predictions, X_test, y_test, clf):
    print("Accuracy: ", accuracy_score(y_test, predictions))
    print('Precision: ', precision_score(y_test, predictions))
    print('Recall: ', recall_score(y_test, predictions))
    print('f1 micro: ', f1_score(y_test, predictions, average='micro'))
    print('f1 macro: ', f1_score(y_test, predictions, average='macro'))
    y_proba = clf.predict(X_test)
    print('ROC AUC: ', roc_auc_score(y_test, y_proba))
    print('Confusion matrix:\n', pd.DataFrame(confusion_matrix(y_test, predictions)))
    plot_roc_curve(clf, X_test, y_test)
    plt.show() 
    
    plt.figure(figsize=(15,3))
    plt.plot(predictions[:300], 'r-', label='Predicted')
    plt.plot(y_test.values[:300], 'b-', label='True')
    plt.show()
```

```python
import warnings
from sklearn.model_selection import GridSearchCV
from sklearn.exceptions import ConvergenceWarning


metrics_cls = ['f1_macro']
def tune_params_classifier(model, parameters, splits, X_train, y_train, metrics):
    for metric in metrics:
        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", category=ConvergenceWarning)
            classifier = GridSearchCV(model, parameters, scoring=metric, cv=splits)
            classifier.fit(X_train, y_train)
            print('****'*10)
            print('Metric:', metric)
            print('Best score:', classifier.best_score_)
            print('Best params:', classifier.best_params_)
```

```python
def fit_eval_classifier(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    evalulate_classifier(predictions, X_test, y_test, clf)
```

```python
import tensorflow as tf
import math
from tensorflow import keras
from sklearn.utils import resample
from tensorflow.keras.models import Sequential
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_circles
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix



from tensorflow.keras.layers import Layer
from tensorflow.keras import backend as K

def f1(y_true, y_pred):
    def recall(y_true, y_pred):
        """Recall metric.

        Only computes a batch-wise average of recall.

        Computes the recall, a metric for multi-label classification of
        how many relevant items are selected.
        """
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

    def precision(y_true, y_pred):
        """Precision metric.

        Only computes a batch-wise average of precision.

        Computes the precision, a metric for multi-label classification of
        how many selected items are relevant.
        """
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision
    precision = precision(y_true, y_pred)
    recall = recall(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))


# –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
numpy.random.seed(2)

model = Sequential()
model.add(Dense(75, input_dim=9, activation='sigmoid'))
model.add(Dense(75, activation='relu'))
model.add(Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(0.01)))

# –∫–æ–º–ø–∏–ª–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å, –∏—Å–ø–æ–ª—å–∑—É–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ adam
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['binary_accuracy'])
print(model.summary())

    
# –æ–±—É—á–∞–µ–º –Ω–µ–π—Ä–æ–Ω–Ω—É—é —Å–µ—Ç—å
model.fit(X_train, y_train, epochs = 10, batch_size=64)

features_test = X_test
labels_test = y_test

predictions = model.predict_classes(features_test)
probs = model.predict(features_test)
predictions = predictions[:, 0]
probs = probs[:, 0]

tn = ['Normal', 'Fake']
matrix = confusion_matrix(labels_test, predictions)
print(matrix)
print(classification_report(labels_test, predictions, target_names=tn))
print('F1 macro', f1_score(labels_test, predictions, average='macro'))
```

```python
import kerastuner
from kerastuner.tuners import RandomSearch, Hyperband, BayesianOptimization
from tensorflow import keras

def build_model(hp):
    model = keras.Sequential()
    # activation_choice = hp.Choice('activation', values=['relu', 'sigmoid', 'tanh', 'elu', 'selu'])    
    model.add(keras.layers.Dense(units=hp.Int('units_input',
                                   min_value=9,
                                   max_value=100,
                                   step=50),
                    input_dim=X_train.shape[1],
                    activation='sigmoid'))
    model.add(keras.layers.Dense(units=hp.Int('units_hidden',        
                                   min_value=2,   
                                   max_value=100,   
                                   step=1),
                    activation='relu'))   
    model.add(keras.layers.Dense(1, activation = "sigmoid"))
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=[f1])
    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective=kerastuner.Objective("val_f1", direction="max"),
    max_trials=80,
    directory='.models'
    )
```
