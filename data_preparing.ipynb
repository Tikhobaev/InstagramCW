```python
import pandas as pd

pd.set_option('display.max_columns', 40)
# pd.set_option('display.max_columns', 10)
pd.set_option('display.max_colwidth', -1)
data = pd.read_csv('data_13k.csv', sep=';')
data = data.drop(data.columns[[0]], axis=1)  # delete index column
data
```

```python
normal = data.loc[data['label'] == 'NORMAL']
print('Number of normal users =', len(normal))

business = data.loc[data['label'] == 'BUSINESS']
print('Number of business users =', len(business))

media = data.loc[data['label'] == 'MEDIA']
print('Number of media users =', len(media))

spammers = data.loc[data['label'] == 'SPAMMER']
print('Number of spammers =', len(spammers))

fakes = data.loc[data['label'] == 'FAKE']
print('Number of fakes =', len(fakes))
```

```python
print('****' * 10, 'Overall number of NaNs:', '****' * 10)
print(data.isnull().sum().sum())
print('\n', '****' * 10, 'Number of NaNs divided by columns:', '****' * 10)
print(data.isnull().sum())
```

```python
prepared_data = data.copy(deep=True)
prepared_data['name'] = prepared_data['name'].fillna('')
prepared_data['biography'] = prepared_data['biography'].fillna('')
prepared_data
```

```python
print('****' * 10, 'Overall number of NaNs:', '****' * 10)
print(prepared_data.isnull().sum().sum())
print('\n', '****' * 10, 'Number of NaNs divided by columns:', '****' * 10)
print(prepared_data.isnull().sum())
prepared_data["isprivate"] = prepared_data["isprivate"].astype(int)
```

```python
print('****' * 10, 'Column types', '****' * 10)
print(prepared_data.dtypes)
for i, row in prepared_data.iterrows():
    prepared_data.at[i,'follower_count'] = prepared_data.at[i,'follower_count'].replace('—Ç—ã—Å.', '000')
    prepared_data.at[i,'follower_count'] = prepared_data.at[i,'follower_count'].replace('–º–ª–Ω', '000000')
    if ',' in prepared_data.at[i,'follower_count']:
        prepared_data.at[i,'follower_count'] = prepared_data.at[i,'follower_count'].replace(',', '')[:-1]
prepared_data["follower_count"] = prepared_data["follower_count"].astype(int)
```

```python
print('****' * 10, 'Column types', '****' * 10)
print(prepared_data.dtypes)
```

```python
import re

prepared_data['bio_len'] = [0]*len(prepared_data)
prepared_data['name_len'] = [0]*len(prepared_data)
prepared_data['bio_word_count'] = [0]*len(prepared_data)
prepared_data['name_word_count'] = [0]*len(prepared_data)
prepared_data['have_com_links'] = [0]*len(prepared_data)
prepared_data['have_ru_links'] = [0]*len(prepared_data)
prepared_data['bio_has_digits'] = [0]*len(prepared_data)
prepared_data['name_has_digits'] = [0]*len(prepared_data)
prepared_data['proportion'] = [0.]*len(prepared_data)
prepared_data['digit_count_username'] = [0]*len(prepared_data)

eps = 0.1

for i, row in prepared_data.iterrows():
    prepared_data.at[i,'biography'] = re.sub(r'–ü–æ–¥–ø–∏—Å–∞–Ω.*', '', prepared_data.at[i,'biography'])
    prepared_data.at[i,'bio_len'] = len(prepared_data.at[i,'biography'])
    prepared_data.at[i,'name_len'] = len(prepared_data.at[i,'name'])
    prepared_data.at[i,'bio_word_count'] = len(prepared_data.at[i,'biography'].split())
    prepared_data.at[i,'name_word_count'] = len(prepared_data.at[i,'name'].split())
    prepared_data.at[i,'isprivate'] = 1 if prepared_data.at[i,'isprivate'] == True else 0
    prepared_data.at[i,'proportion'] = prepared_data.at[i,'following_count'] / (prepared_data.at[i,'follower_count'] + eps)
    prepared_data.at[i,'digit_count_username'] = len(re.findall(r'\d', prepared_data.at[i,'username']))
    if '.com' in str(prepared_data.at[i,'biography']):
        prepared_data.at[i,'have_com_links'] = 1
    if '.ru' in str(prepared_data.at[i,'biography']):
        prepared_data.at[i,'have_ru_links'] = 1
    # digits in bio
    res = re.search(r'\d+', str(prepared_data.at[i,'biography']))
    if res:
        prepared_data.at[i,'bio_has_digits'] = 1
    # digits in name
    res = re.search(r'\d+', str(prepared_data.at[i,'name']))
    if res:
        prepared_data.at[i,'name_has_digits'] = 1

prepared_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


def get_numeric_columns(data):
    numeric_columns = []
    for column in data:
        if data[column].dtype != 'object':
            numeric_columns.append(column)
    return numeric_columns
        
numeric_columns = get_numeric_columns(prepared_data)
print(numeric_columns)
```

```python
for column in numeric_columns:
    plt.figure(figsize=(15,10))
    sns.stripplot(x='label', y=column, data=prepared_data, jitter=True)
    plt.show()
```

```python
features = prepared_data.copy(deep=True)
labels = features['label']
features = features[get_numeric_columns(prepared_data)]

labels
```
